```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('datasets/Position_Salaries.csv')
X = dataset.iloc[:, 1:-1]
y = dataset.iloc[:, -1]

print('end')
```

```python
from sklearn.model_selection import train_test_split

# x_train, x_test, y_train, y_test = train_test_split(
#     X, y, test_size = 0.2, random_state = 0)

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)

print('done')
```

```python
from sklearn.preprocessing import PolynomialFeatures

polynomBuilder = PolynomialFeatures(degree=4)
X_polynom = polynomBuilder.fit_transform(X)

lr_polynom = LinearRegression()
lr_polynom.fit(X_polynom, y)
predicted = lr_polynom.predict(X_polynom)

print('done')
```

```python
import matplotlib.pyplot as plot

plot.scatter(X, y, color='red')
plot.plot(X, lr.predict(X), color='blue')
plot.plot(X, predicted, color='green')

print('done')
```

### Making the curve look smoother by generating intermediate X-axis coordinates


```python
X_col0 = X.iloc[:, 0]
X_hd = np.arange(min(X_col0), max(X_col0), step=0.1)
X_hd = X_hd.reshape((len(X_hd), 1))
X_hd_poly = polynomBuilder.fit_transform(X_hd)
predicted_hd = lr_polynom.predict(X_hd_poly)
plot.scatter(X, y, color='red')
plot.plot(X_hd, predicted_hd, color='green')
print('done')
```

```python
# Predicting y for arbitrary x==6.5

target_job_role = [[6.5]]
X1_poly = polynomBuilder.fit_transform(target_job_role)
target_salary = lr_polynom.predict(X1_poly)
print(target_salary)

plot.scatter(X, y, color='red')
plot.plot(X_hd, predicted_hd, color='green')
plot.scatter(target_job_role, target_salary, color='black')

print('done')
```
