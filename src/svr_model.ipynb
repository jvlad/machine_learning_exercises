```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
from sklearn.preprocessing import StandardScaler
from numpy.testing import assert_allclose
```

```python
dataset = pd.read_csv("datasets/Position_Salaries.csv")
X = dataset.iloc[:, 1:-1]
y = dataset.iloc[:, -1]

expected_x_range = np.arange(1,11).reshape(10,1)
assert_allclose(X, expected_x_range)

print('end')
```

```python
# scaling features
scalerX = StandardScaler()
scalerY = StandardScaler()
X_scaled = scalerX.fit_transform(X)
y_scaled = scalerY.fit_transform(y.values
                                .reshape((len(y), 1))).reshape((len(y)))

assert_allclose(y_scaled, [-0.72004253, -0.70243757, -0.66722767, -0.59680786, -0.49117815,-0.35033854, -0.17428902,  0.17781001,  0.88200808,  2.64250325])
print('end')
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel='rbf')
regressor.fit(X_scaled, y_scaled)

import matplotlib.pyplot as plot

plot.scatter(X_scaled, y_scaled, color='red')
predicted_scaled = regressor.predict(X_scaled)
plot.plot(X_scaled, predicted_scaled, color='blue')

print('done')
```

```python
plot.scatter(X, y, color='red')
X_unscaled = scalerX.inverse_transform(X_scaled)
predicted_unscaled = scalerY.inverse_transform([predicted_scaled])
plot.plot(X, predicted_unscaled[0], color='blue')


target_job_role = 6.5
target_job_role_scaled = scalerX.transform([[target_job_role]])
target_salary_scaled = regressor.predict(target_job_role_scaled)
target_salary = scalerY.inverse_transform([target_salary_scaled])
plot.scatter(target_job_role, target_salary, color='blue')
print(target_salary)
reformat_y_axis_ticks(plot)

def reformat_y_axis_ticks(plot):  
  current_values = plot.gca().get_yticks()
  plot.gca().set_yticklabels(['{:.0f}'.format(x) for x in current_values])

print('done')
```
